**2. What pokemon has the ID of 55?**

```python
import requests
response = requests.get("https://pokeapi.co/api/v2/pokemon/55/")
pokemon_55_data = response.json()
(f"The pokemon with an ID of 55 is {pokemon_55_data['name']}.")
```

**3. How tall is that pokemon?**

```python
(f"The Pokemon {pokemon_55_data['name']} has a height of {pokemon_55_data['height']}.")
```

**4. How many versions of Pokemon games have been released?**

```python
response = requests.get("https://pokeapi.co/api/v2/version/")
pokemon_versions = response.json()
pokemon_versions['count']
```

**5. Print out the name of every electric-type pokemon.**

```python
response = requests.get("https://pokeapi.co/api/v2/type/electric/")
electric_pokemon = response.json()

[ind_pokemon['pokemon']['name'] for ind_pokemon in electric_pokemon['pokemon']]
```

**6. What are electric-type Pokemon called in the Korean version of the game? (i.e. what do they call electric-type pokemon in Korean if not "electric")**

```python
[electric_lang['name'] for electric_lang in electric_pokemon['names'] if electric_lang['language']['name'] == "ko"]
```

**7. Who has a higher speed stat, Eevee or Pikachu?**

```python
response = requests.get("https://pokeapi.co/api/v2/pokemon/pikachu/")
pikachu = response.json()

pikachu_speed = [speed['base_stat'] for speed in pikachu['stats'] if speed['stat']['name'] == "speed"]
pikachu_speed

response = requests.get("https://pokeapi.co/api/v2/pokemon/eevee/")
eevee = response.json()

eevee_speed = [speed['base_stat'] for speed in eevee['stats'] if speed['stat']['name'] == "speed"]
eevee_speed

if eevee_speed > pikachu_speed:
    print(f"Eevee (speed={eevee_speed}) is faster than Pikachu (speed: {pikachu_speed} .")
elif pikachu_speed > eevee_speed:
    print(f"Pikachu (speed: {pikachu_speed}) is faster than Eevee (speed: {eevee_speed}).")
```
