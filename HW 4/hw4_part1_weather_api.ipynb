**2. Make a request for the current weather where you are born, or somewhere you've lived.**

```python
import requests
from dotenv import load_dotenv
import os

load_dotenv()
weather_api_key=os.getenv("weather_api_key")

response = requests.get(f"http://api.weatherapi.com/v1/current.json?key={weather_api_key}&q=Chicago")
chicago_data = response.json()
chicago_data
```

**3. Print out the country this location is in.**

```python
chicago_data['location']['country']
```

**4. Print out the difference between the current temperature and how warm it feels. Use "It feels ___ degrees colder" or "It feels ___ degrees warmer," not negative numbers.**

```python
print(f"The current actual temperature is {chicago_data['current']['temp_f']:.1f} degrees F.")
print(f"The current feels-like temperature is {chicago_data['current']['feelslike_f']:.1f} degrees F.")

if chicago_data['current']['feelslike_f'] > chicago_data['current']['temp_f']:
    print(f"It feels {chicago_data['current']['feelslike_f'] - chicago_data['current']['temp_f']:.1f} degrees warmer")
elif chicago_data['current']['temp_f'] > chicago_data['current']['feelslike_f']:
    print(f"It feels {chicago_data['current']['temp_f'] - chicago_data['current']['feelslike_f']:.1f} degrees colder")
elif chicago_data['current']['temp_f'] == chicago_data['current']['feelslike_f']:
    print("The feels-like and actual temperatures are the same.")
```
